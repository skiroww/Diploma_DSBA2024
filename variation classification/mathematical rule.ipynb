```python
import pandas as pd
import numpy as np
from scipy.stats import wilcoxon, levene
from navec import Navec
from slovnet.model.emb import NavecEmbedding
import torch 
import ast
```

```python
path = r'C:\Users\User\Documents\github\He_who_laughs_last\data\embedded\navec.tar'  
navec = Navec.load(path)
 
```

```python
# получаем эмбединги
navec['слово']
```

```python
#загружаем шутки)
dataframe_jokes_russian  = pd.read_csv(r'C:\Users\User\Documents\github\He_who_laughs_last\Data_preprocessing_classification\data\lemmatized_russian.csv').dropna()
```

```python
column_of_lists = [ast.literal_eval(x) for x in dataframe_jokes_russian['lemmatized']]
```

```python
dataframe_jokes_russian['lists'] = column_of_lists
```

```python
dataframe_jokes_russian
```

```python
embedded_big_list = []
for joke in dataframe_jokes_russian['lists']:
    joke_navec = []
    for word in joke:
        try:

            joke_navec.append(navec[word])
        except:
            joke_navec.append(navec['<unk>'])
    embedded_big_list.append(joke_navec)
```

```python
dataframe_jokes_russian['navec'] = embedded_big_list
```

```python
variances_dataset = []
for emb in dataframe_jokes_russian['navec']:
    #print(len(emb))
    variances_dataset.append(np.var(emb))
```

```python
dataframe_jokes_russian['variances'] =variances_dataset
```

```python
np.median(dataframe_jokes_russian['variances'])
```
